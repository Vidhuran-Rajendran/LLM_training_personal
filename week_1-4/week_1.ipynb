```python
nums = [1,1,1,2,2,3]
k = 2
import heapq
from collections import Counter
heap=[]
print("heap",heap)
count = Counter(nums)
print(count)
for key,val in count.items():
    if len(heap) < k :
        heapq.heappush(heap,(val,key))
        print(heap)
    else:
        heapq.heappushpop(heap,(val,key))
        print("heappop",heap)
print([h[1]for h in heap])
```

```python
chat_history=[]
cache = {}
user_input = str
system = "your are a Strict DSA tutor always explain in 3 step like what is the concept, real world example and code example in python"
chat_history.append({"role":"system","content":system,})

def chatbot(chat_history,cache):
    try:
        response = ollama.chat(model="qwen2.5",
                        messages =chat_history)
    except Exception as e:
        print("Model Error")
        return
    assistance_responce = response["message"]['content']
    chat_history.append({"role":"assistant","content":assistance_responce})
    last_user = chat_history[-2]["content"]
    cache[last_user] = assistance_responce
    print(assistance_responce)

while True:
    user_input = input("chat: ").strip()
    if user_input.lower()=='quit':
        break
    if not user_input:
        print("Empty message")
        
    chat_history.append({"role":"user","content":user_input})

    if user_input in cache:
        chat_history.append({"role":"assistant","content":cache[user_input]})

        print("\n(From Cache):", cache[user_input], "\n")
    else:
        chatbot(chat_history,cache)
```

```python
import ollama
from collections import OrderedDict

#config section
model = "qwen2.5"
system = "your are a Strict DSA tutor always explain in 3 step like what is the concept, real world example and code example in python"

#memort section
chat_history = []
cache = OrderedDict()
ollama_counter = 0
cache_counter = 0

#logic section
def chatbot(chat_history,model):
    try:
        response = ollama.chat(model=model,messages=chat_history) # initiating the model
    except Exception as e:
        print("Model not found")
        return
    assistance_responce = response["message"]["content"] # initiating the models responce to the user input
    chat_history.append({"role":"assistant","content":assistance_responce}) # storing in the history
    return assistance_responce
    
    
#interface Section

while True:
    user_input = input("chat: ").strip()
    
    if  user_input.lower() == "quit":
        print('ollama_counter : ',ollama_counter)
        print('cache_counter : ',cache_counter)
        break
    
    chat_history.append({"role":"user","content":user_input})
    
    if user_input in cache:
        print(cache[user_input])
        chat_history.append({"role":"assistant","content":cache[user_input]})
        cache_counter+=1
    else:  
        ai_reply = chatbot(chat_history,model)
        ollama_counter+=1
        if len(cache)>=5:
            cache.popitem(last=False)
        if ai_reply!=None:
            cache[user_input] = ai_reply
            print(ai_reply)
```
