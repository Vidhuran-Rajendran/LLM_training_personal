```python
import pymupdf

def convert_pdf_to_txt(pdf_path, output_path):
    # Open the document
    doc = pymupdf.open(pdf_path)
    full_text = ""

    # Iterate through pages and extract text
    for page in doc:
        full_text += page.get_text() + "\n"

    # Save to a text file
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(full_text)
    
    print(f"Success! Saved to {output_path}")

# Usage
convert_pdf_to_txt("E:\\training\\LLM_training\\Dsa.pdf", "output.txt")
```

## Basic pdf to md conversion

```python
import pymupdf4llm
import pathlib

# Define your input and output paths
input_pdf = "E:\\training\\LLM_training\\Dsa.pdf"
output_md = "output.md"

# 1. Convert PDF to Markdown text
md_text = pymupdf4llm.to_markdown(input_pdf)

# 2. Save the output to a file
pathlib.Path(output_md).write_bytes(md_text.encode())

print(f"Successfully converted {input_pdf} to {output_md}")
```

## Advance conversion with removal

```python
import pymupdf4llm
import pathlib
import re

# ----------- FILE PATHS -----------
input_pdf = "E:\\training\\LLM_training\\Dsa.pdf"
output_md = "clean_output.md"


# ----------- CLEANING FUNCTION -----------
def clean_text(text):

    # ✅ 1. Remove Chapter 1 (Introduction)
    text = re.sub(r"Chapter 1.*?Chapter 2", "Chapter 2", text, flags=re.DOTALL)

    # ✅ 2. Remove Appendices (A–E)
    text = re.sub(r"Appendix A.*", "", text, flags=re.DOTALL)

    # ✅ 3. Remove Table of contents (if exists)
    text = re.sub(r"Contents.*?Chapter 2", "Chapter 2", text, flags=re.DOTALL)

    # ✅ 4. Remove image-only lines
    text = re.sub(r"!\[\]\[.*?\]", "", text)

    # ✅ 5. Remove weird symbols
    text = re.sub(r"[^\x00-\x7F]+", " ", text)

    # ✅ 6. Remove numbering like 1) 2)
    text = re.sub(r"\b\d+\)", "", text)

    # ✅ 7. Remove 'algorithm' blocks (pseudo code)
    text = re.sub(r"algorithm.*?end\s+\w+", "", text, flags=re.DOTALL | re.IGNORECASE)

    # ✅ 8. Remove CHAPTER headers
    text = re.sub(r"_CHAPTER.*?\n", "", text)

    # ✅ 9. Keep only important chapters (2–7)
    text = re.sub(r"Chapter 8.*", "", text, flags=re.DOTALL)

    # ✅ 10. Normalize spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()


# ----------- PROCESS -----------
print("Converting PDF → Markdown...")
md_text = pymupdf4llm.to_markdown(input_pdf)

print("Cleaning text...")
cleaned_text = clean_text(md_text)

print("Saving cleaned .md file...")
pathlib.Path(output_md).write_text(cleaned_text, encoding="utf-8")

print(f"✅ Clean Markdown saved to: {output_md}")
```
